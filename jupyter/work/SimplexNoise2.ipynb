```python
import sys
import math

from PIL import Image

import numpy as np
from bisect import bisect_left

import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from matplotlib import cm

# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D

from pprint import pprint
```

```python
# https://github.com/lmas/opensimplex/
# https://github.com/lmas/opensimplex/blob/master/example/images.py

import open_simplex as simplex
```

```python
WIDTH = 256
HEIGHT = 256
FEATURE_SIZE = 24.0
```

```python
print('Generating 2D image...')
im = Image.new('L', (WIDTH, HEIGHT))
for y in range(0, HEIGHT):
    for x in range(0, WIDTH):
        value = simplex.noise2(x / FEATURE_SIZE, y / FEATURE_SIZE)
        color = int((value + 1) * 128)
        im.putpixel((x, y), color)
#im.save('noise2d.png')

plt.imshow(im)
plt.show()
```

```python
print('Generating 2D slice of 3D...')
im = Image.new('L', (WIDTH, HEIGHT))
for y in range(0, HEIGHT):
    for x in range(0, WIDTH):
        value = simplex.noise3(x / FEATURE_SIZE, y / FEATURE_SIZE, 0.0)
        color = int((value + 1) * 128)
        im.putpixel((x, y), color)
#im.save('noise3d.png')

plt.imshow(im)
plt.show()
```

```python
print('Generating 2D slice of 4D...')
im = Image.new('L', (WIDTH, HEIGHT))
for y in range(0, HEIGHT):
    for x in range(0, WIDTH):
        value = simplex.noise4(x / FEATURE_SIZE, y / FEATURE_SIZE, 0.0, 0.0)
        color = int((value + 1) * 128)
        im.putpixel((x, y), color)
#im.save('noise4d.png')

plt.imshow(im)
plt.show()
```
